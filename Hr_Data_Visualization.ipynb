First, we must import the necessary packages to create our notebook. For this project, I have imported csv, pandas, numpy, matplotlib, and seaborn.

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mtplt
import seaborn as sns
```

Next, we must import our data.  Once our data is imported, we can begin to manipulate it as we see fit.

```python
data = pd.read_csv('HR_comma_sep.csv')
data.head()
```

```python
data.tail()
```

```python
count_salary = pd.value_counts(data['salary'], sort = True).sort_index()
count_salary
```

For our first visualization, I will take a count of each salary and use a pie chart to display what type of salary an employee is most likely to have, high, medium, or low.

```python
count_salary.plot(kind = "pie", legend = True)
plt.show()
```

```python
data.columns
```

```python
df1 = data.set_index('salary')
df1 = df1.groupby('salary').mean().sort_values('satisfaction_level',ascending=False)
df1
```

For my next visualization, I wanted to see some of the correlations between the salaries and other key pieces of data.  To show this, I created three vertical bar charts.  The first, shows the satisfaction level as it relates to the employees salary.  the next, shows how many monthly hours they bill.  The last shows the time spent at the company.

```python
df1.plot(kind='bar', y='satisfaction_level', title = 'Satisfaction Level', legend = False)
df1.plot(kind='bar', y='average_montly_hours', color = 'green', title = 'Average Monthly Hours', legend = False)
df1.plot(kind='bar', y='time_spend_company', color = 'yellow', title = 'Time Spent at the Company', legend = False)
plt.show()
```

Next, I wanted to see some correlations between satisfaction levels at a given company.  To display this, I chose to use a KDE plot, a histogram, and a scatter plot.

```python
sns.jointplot(x="satisfaction_level", y="time_spend_company", data=df1, kind="kde")
plt.show()
```

```python
data.plot(kind='hist', x='time_spend_company', y='satisfaction_level', color = 'pink')
plt.show()
```

```python
df3 = data
sns.jointplot(x="last_evaluation", y="satisfaction_level", data=df3, size = 12)
plt.show()
```
